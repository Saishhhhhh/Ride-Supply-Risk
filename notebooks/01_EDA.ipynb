```python
df['Booking Value'] = df.groupby('Vehicle Type')['Booking Value'].transform(
        lambda x: x.fillna(x.median())
    )

df['Ride Distance'] = df['Ride Distance'].fillna(
        df.groupby('Vehicle Type')['Ride Distance'].transform('median')
    )
```

```python
# Dropping Columns
columns_to_drop =  ['Booking Status', 'Reason for cancelling by Customer', 'Driver Cancellation Reason', 'Incomplete Rides Reason', 'Customer ID', 'Booking ID', 'Avg CTAT']
df.drop(columns_to_drop , axis = 1, inplace=True)
df.head(5)
```

```python
df['Cancelled Rides by Customer'].fillna(0, inplace=True)
```

```python
df['Cancelled Rides by Driver'].fillna(0, inplace=True)
```

```python
df['Incomplete Rides'].fillna(0, inplace=True)
```

```python
df.sample(20)
```

```python
df.isnull().sum()
```

```python
#Drive Rating Fix
df['driver_rating_missing'] = df['Driver Ratings'].isnull().astype(int)
driver_median = df['Driver Ratings'].median()
df['driver_rating_filled'] = df['Driver Ratings'].fillna(driver_median)
```

```python
#Customer Rating Fix
df['customer_rating_missing'] = df['Customer Rating'].isnull().astype(int)
customer_median = df['Customer Rating'].median()
df['customer_rating_filled'] = df['Customer Rating'].fillna(customer_median)
```

```python
#Payment Fix
df['Payment Method'] = df['Payment Method'].fillna('UNKNOWN')
```

```python
# Dropping Columns
columns_to_drop =  ['Customer Rating', 'Driver Ratings']
df.drop(columns_to_drop , axis = 1, inplace=True)
df.head(5)
```

```python
df['Supply_Stress'] = (
    (df['Avg VTAT'].notna())) & (df["Avg VTAT"] > df["Avg VTAT"].quantile(0.90)).astype(int)
```

```python
df = df.drop(columns=["Avg VTAT"])
```

```python
df["hour"] = pd.to_datetime(df["Time"]).dt.hour
df["day_of_week"] = pd.to_datetime(df["Date"]).dt.dayofweek
```

```python
df = df.drop(columns=["Date", "Time"])
```

```python
df["Supply_Stress"] = df["Supply_Stress"].astype(int)
```

```python
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
df.drop(['hour', 'day_of_week'], axis=1, inplace=True)
```

```python
df.info()
```

```python
categorical_cols = ['Vehicle Type', 'Pickup Location', 'Drop Location', 'Payment Method']
encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le
```
